```python
import numpy as np
n=12
np.random.seed(488)
a = np.random.randint(2, size=(n, n))
a
```

```python
b=[]
count=0
island=0
print(a)
#所有1的位置
for i in range(n):
    for j in range(n):
        if a[i,j]==1:
            b.append([j,i])
print(b)


# 判斷
for c in b:
    #buffer歸零
    buffer=[0]
    
    #如果checked(2),就跳迴圈
    if a[c[1],c[0]]==2:
        continue
    
    #計算island數量
    print("=====")
    island+=1
    print("第",island,"座島")
    
    #初步尋找矩陣值不為零的座標
    if c[0]+1!=n: #邊界條件
        if a[c[1],c[0]+1]==1: # x+1 位置判斷(往右一格)
            a[c[1],c[0]+1]=3
            buffer.append([c[0]+1,c[1]])
            count+=1
    if c[0]-1!=-1:
        if a[c[1],c[0]-1]==1: # x-1 位置判斷(往左一格)
            a[c[1],c[0]-1]=3
            buffer.append([c[0]-1,c[1]])
            count+=1
    if c[1]+1!=n:
        if a[c[1]+1,c[0]]==1: # y+1 位置判斷(往下一格)
            a[c[1]+1,c[0]]=3
            buffer.append([c[0],c[1]+1])
            count+=1
    if c[1]-1!=-1:
        if a[c[1]-1,c[0]]==1: # y-1 位置判斷(往上一格)
            a[c[1]-1,c[0]]=3
            buffer.append([c[0],c[1]-1])
            count+=1
    
    #將第一個不為零的矩陣位置改為checked並印出
    a[c[1],c[0]]=2
    print(a) #目前狀況
    print(buffer, count)
    
    #增加buffer格數並印出buffer
    for l in range(count,n+5):
        buffer.append([0,0])
    print(buffer, count)

    #buffer功能
    while buffer[count]!=0:
        #目前判斷的位置
        print("位置: ",buffer[count])
        #偵測到1的數量
        add=0
        if buffer[count][0]+1!=n: #邊界條件
            if a[buffer[count][1],buffer[count][0]+1]==1: # x+1 位置判斷(往右一格)
                a[buffer[count][1],buffer[count][0]+1]=3 #將值命為pulled(3)
                add+=1
                buffer[count+add]=[buffer[count][0]+1,buffer[count][1]] #往buffer後一格放
                print("---判斷---增加: ",[buffer[count][0]+1,buffer[count][1]])
                print(buffer, count)
        if buffer[count][0]-1!=-1:
            if a[buffer[count][1],buffer[count][0]-1]==1: # x-1 位置判斷(往左一格)
                a[buffer[count][1],buffer[count][0]-1]=3
                add+=1
                buffer[count+add]=[buffer[count][0]-1,buffer[count][1]]
                print("---判斷---增加: ",[buffer[count][0]-1,buffer[count][1]])
                print(buffer, count)
        if buffer[count][1]+1!=n:
            if a[buffer[count][1]+1,buffer[count][0]]==1: # y+1 位置判斷(往下一格)
                a[buffer[count][1]+1,buffer[count][0]]=3
                add+=1
                buffer[count+add]=[buffer[count][0],buffer[count][1]+1]
                print("---判斷---增加: ",[buffer[count][0],buffer[count][1]+1])
                print(buffer, count)
        if buffer[count][1]-1!=-1:
            if a[buffer[count][1]-1,buffer[count][0]]==1: # y-1 位置判斷(往上一格)
                a[buffer[count][1]-1,buffer[count][0]]=3
                add+=1
                buffer[count+add]=[buffer[count][0],buffer[count][1]-1]
                print("---判斷---增加: ",[buffer[count][0],buffer[count][1]-1])
                print(buffer, count)
            
        a[buffer[count][1],buffer[count][0]]=2 #將目前的值令checked(2)
        print("---移位---")
        for k in range(count,count+add+1): #將buffer座標全部往前移
            buffer[k]=buffer[k+1]
        count=count+add-1 #接下來要測的座標
        print(buffer, count)
        print(a)
        print("=====")

print("共",island,"座島")
```
